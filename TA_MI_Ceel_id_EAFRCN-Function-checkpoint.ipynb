# Cleaning Function

```python
def data_cleaning(x):
    x=x[x['Time'].notna()]
    return x
```

# TA Function

```python
def TA(x):
    x['Time']=pd.to_datetime(x.Time)
    tg=pd.DataFrame(x.groupby(pd.Grouper(key='Time',freq='2s',offset='1s')),columns=['time','summary'])
    two_sec=tg['time'].unique()
    two_sec=list(two_sec)
    common = set.intersection(set(x.Time), set(tg.time))
    common_2=list(common)
    common_2.sort()
    
    
    empty_df=pd.DataFrame()
    for j in range(len(common_2)):
        y=x.groupby(pd.Grouper(key='Time',freq='2s',offset='1s')).get_group(common_2[j])
        y['Timing Advance (meters) Carrier 1'].fillna(y['Timing Advance (meters) Carrier 1'].mean(),inplace=True)
        empty_df=pd.concat([empty_df,y])
    return empty_df
```

# MI Function ( EAFRCN)

```python
def MI(x):
    a=x['Time'].iloc[0]
    x=x[x['Serving Cell Identity'].notna()]
    mi=pd.DataFrame(x.groupby(pd.Grouper(key='Time',freq='60s',origin=a)),columns=['time','summary'])
    common_mi= set.intersection(set(x.Time), set(mi.time))
    common_mi_2=list(common_mi)
    common_mi_2.sort()
    MI_df=pd.DataFrame()
     
    for i in range(len(common_mi_2)):
        y=x.groupby(pd.Grouper(key='Time',freq='60s',origin=a)).get_group(common_mi_2[i])
        NaN=np.nan
        y['MI']=NaN
        eafrcn_uni=y['Serving Cell DL EARFCN'].unique()
        EAFRCNs=list(eafrcn_uni)
        MI_EAFRCN=pd.DataFrame()
        for EAFRCN in EAFRCNs:
            counter=0
            z=y.groupby('Serving Cell DL EARFCN').get_group(EAFRCN)
            for j in range(len(z)-1):    
                if z['Serving Cell Identity'].iloc[j]==z['Serving Cell Identity'].iloc[j+1]:
                    counter=counter
                else:
                    counter=counter+1
            z['MI'].fillna(counter,inplace=True)
            MI_EAFRCN=pd.concat([MI_EAFRCN,z]) 
        
        MI_df=pd.concat([MI_df,MI_EAFRCN]) 
    return MI_df
```
